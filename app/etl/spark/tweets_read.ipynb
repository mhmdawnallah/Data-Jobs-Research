# Most popular job titles in data field e.g:(Big Data, Data Science, Data Analaysis) over recent 7 days?

```python
client = tweepy.Client(bearer_token,consumer_key,consumer_secret,access_token,access_token_secret)
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
response = client.search_recent_tweets(query="data engineering hiring", max_results=100, tweet_fields=['created_at','lang'],expansions=['author_id'])
query = "#bigdata"
counts = client.get_recent_tweets_count(query=query)
for count in counts.data:
    print(count)
print("Finished Big Data")
query = "#datascience"
counts = client.get_recent_tweets_count(query=query)
for count in counts.data:
    print(count)
```

```python
class IDPrinter(tweepy.StreamingClient):
    def __init__(self, bearer_token,client_socket):
        self.client_socket = client_socket
        super().__init__(bearer_token)
        
    def on_tweet(self, tweet):
        tweet_message: str = tweet.text
        print(tweet)
        self.client_socket.send(tweet_message.encode('utf-8'))        

    def on_errors(self, errors):
        print(f"Received error code {errors}")
        self.disconnect()
        return False


def send_data(client_socket):
    printer = IDPrinter(bearer_token,client_socket)
    printer.add_rules(tweepy.StreamRule("#bigdata"))
    printer.add_rules(tweepy.StreamRule("#datascience"))
    printer.add_rules(tweepy.StreamRule("#dataanalysis"))
    printer.add_rules(tweepy.StreamRule("#dataanalytics")) 


    printer.filter()
```

```python
if __name__ == "__main__":
    s = socket.socket() 
    host = "127.0.0.1"
    port = 5555
    s.bind((host, port))

    print("Listening on port: %s" % str(port))

    s.listen(5)
    c, addr = s.accept()

    print("Received request from: " + str(addr))

    send_data(c)
```
